## Matplotlib Example Workflow

```python
# 0. Importing matplotlib and getting it ready for plotting in Jupyter Notebook.
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data.
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot.
fig, ax = plt.subplots(figsize=(10,10))   #(width, height)

# 3. Plot data.
ax.plot(x, y)

# 4. Customize plot.
ax.set(title = "Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save and Show (you save the whole figure)
plt.savefig("images\sample-plot.png")
```

## Making figures with NumPy arrays.

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data.
x = np.linspace(0, 10, 100)        # Returns evenly spaces (,,0) samples between start and end.
x[:10]
```

```python
# Plot and create a line plot.
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x, x**2);
```

```python
# Use same data and make a scatter plot.
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x, np.exp(x));
```

```python
# Another scatterplot.
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x, np.sin(x));
```

```python
# Making a plot from dictionary.
nut_butter_prices = {"Almond Butter": 10,
                    "Peanut Butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots(figsize=(10,10))
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Nut-Butter-Store",
      ylabel="Prices");
```

```python
# Horizontal Bar
fig, ax = plt.subplots(figsize=(10,10))
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histogram and plot it.
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

##  Two options for subplot.

```python
# Subplots option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows = 2,
                                          ncols = 2,
                                          figsize=(10,5))

# Plot to each different axes.
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2 (indexing of axes)
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize=(10,10))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(100));
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]", "")
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1,1,2022", periods = len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Let's plot total sales.
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
# Let's try on another dataset.
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Create histogram for age
heart_disease.age.plot.hist(bins = 50);
```

```python
heart_disease.plot.hist(figsize = (10, 30), subplots = True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method.
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = "target");
```

```python
# O-O method mixed with pyplot method.
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = "scatter",
            x = "age",
            y = "chol",
            c = "target", 
            ax = ax);
# ax.set_xlim([45,100]);
```

```python
# O-O method from scratch.
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data.
scatter = ax.scatter(x = over_50.age,
                    y = over_50.chol,
                    c = over_50.target)

# Customize the plot.
ax.set(title = "Heart Disease & Cholestrol Levels",
      xlabel = "Age",
      ylabel = "Cholestrol")

# Add a Legend.
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line.
ax.axhline(over_50.chol.mean(),
          linestyle = "--");
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach.
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                             ncols = 1,
                             figsize = (10, 10))

# Add data to ax0.
scatter = ax0.scatter(x = over_50.age,
                     y = over_50.chol,
                     c = over_50.target)

# Customize ax0.
ax0.set(title = "Heartdisease & Cholestrol levels.",
       xlabel = "Age",
       ylabel = "Cholestrol")

# Add a legent to ax0.
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line.
ax0.axhline(over_50.chol.mean(),
           linestyle = "--")


# Add data to ax1.
scatter2 = ax1.scatter(x = over_50.age,
                      y = over_50.thalach,
                      c = over_50.target)

# Customise ax1.
ax1.set(title = "Heart disease and Thalach level",
        xlabel = "Age",
        ylabel = "Thalach")

# Add a legend.
ax1.legend(*scatter2.legend_elements(),
           title = "Target")

# Add a horizi=ontal line to ax1.
ax1.axhline(over_50.thalach.mean(),
            linestyle = "--")

# Add title to figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```

```python
plt.style.use("seaborn-whitegrid")

# O-O method from scratch.
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data.
scatter = ax.scatter(x = over_50.age,
                    y = over_50.chol,
                    c = over_50.target,
                    cmap = "winter")  # this changes colour

# Customize the plot.
ax.set(title = "Heart Disease & Cholestrol Levels",
      xlabel = "Age",
      ylabel = "Cholestrol")

# Add a Legend.
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line.
ax.axhline(over_50.chol.mean(),
          linestyle = "--");
```

```python
# Customizing y and x axis limitations.

# Subplot of chol, age, thalach.
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                             ncols = 1,
                             figsize = (10, 10))

# Add data to ax0.
scatter = ax0.scatter(x = over_50.age,
                     y = over_50.chol,
                     c = over_50.target,
                     cmap = "winter")

# Customize ax0.
ax0.set(title = "Heartdisease & Cholestrol levels.",
       xlabel = "Age",
       ylabel = "Cholestrol")

# Change x-axis limits.
ax0.set_xlim([50, 80])

# Add a legent to ax0.
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line.
ax0.axhline(over_50.chol.mean(),
           linestyle = "--")


# Add data to ax1.
scatter2 = ax1.scatter(x = over_50.age,
                      y = over_50.thalach,
                      c = over_50.target,
                      cmap = "winter")

# Customise ax1.
ax1.set(title = "Heart disease and Thalach level",
        xlabel = "Age",
        ylabel = "Thalach")

# Change x-axis limits.
ax1.set_xlim([50, 80])

# Change y-axis limits.
ax1.set_ylim([60, 200])

# Add a legend.
ax1.legend(*scatter2.legend_elements(),
           title = "Target")

# Add a horizi=ontal line to ax1.
ax1.axhline(over_50.thalach.mean(),
            linestyle = "--")

# Add title to figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
